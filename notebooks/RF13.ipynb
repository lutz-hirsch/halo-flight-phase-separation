```python
takeoff = 0
touchdown = len(bahamas['time']) - 1

#1st circle set
###############


enter_c1_ds = fpt.ds_timestamp_ind(bahamas, dropsondes, 0)
enter_c1 = fpt.timestamp_ind_1min_prior(bahamas, enter_c1_ds)
exit_c1 = fpt.exit_circle_timestamp_ind(bahamas, enter_c1)
#enter_c1 = fpt.ds_timestamp_ind(bahamas, dropsondes,  0)
#exit_c1  = fpt.ds_timestamp_ind(bahamas, dropsondes, 12)
print ('c1')
print ('----')
print (bahamas.heading[enter_c1_ds])
print ('----')
print (exit_c1)
print (bahamas.heading[exit_c1])
print ('----')
exit_c1 = 4829
print (exit_c1)
print (bahamas.heading[exit_c1])
print ('----')

enter_c2_ds = fpt.ds_timestamp_ind(bahamas, dropsondes, 13)
enter_c2 = fpt.timestamp_ind_1min_prior(bahamas, enter_c2_ds)
exit_c2 = fpt.exit_circle_timestamp_ind(bahamas, enter_c2)
#enter_c2 = fpt.ds_timestamp_ind(bahamas, dropsondes, 13)
#exit_c2  = fpt.ds_timestamp_ind(bahamas, dropsondes, 23)
print ('c2')
print ('----')
print (bahamas.heading[enter_c2_ds])
print ('----')
print (exit_c2)
print (bahamas.heading[exit_c2])
print ('----')
exit_c2 = 8998
print (exit_c2)
print (bahamas.heading[exit_c2])
print ('----')

enter_c3_ds = fpt.ds_timestamp_ind(bahamas, dropsondes, 24)
enter_c3 = fpt.timestamp_ind_1min_prior(bahamas, enter_c3_ds)
exit_c3 = fpt.exit_circle_timestamp_ind(bahamas, enter_c3)
#enter_c3 = fpt.ds_timestamp_ind(bahamas, dropsondes, 24)
#exit_c3  = fpt.ds_timestamp_ind(bahamas, dropsondes, 35)
print ('c3')
print ('----')
print (bahamas.heading[enter_c3_ds])
print ('----')
print (exit_c3)
print (bahamas.heading[exit_c3])
print ('----')
exit_c3 = 12940
print (exit_c3)
print (bahamas.heading[exit_c3])
print ('----')
headingdiff_c3 = bahamas['heading'][exit_c3]-bahamas['heading'][enter_c3]+360


#leg1 (towards NTAS)
min_roll_before_leg1 = fpt.min_timestamp_ind(bahamas, 'roll', 
                       slice('2020-02-13T11:30:00', '2020-02-13T11:45:00'), order=1000)

enter_leg1 = fpt.find_first_value_after_ts_ind(bahamas, 'roll', min_roll_before_leg1, 0., 0.05)

min_roll_after_leg1 = fpt.min_timestamp_ind(bahamas, 'roll', 
                      slice('2020-02-13T12:00:00', '2020-02-13T12:15:00'), order=1000)

exit_leg1 = fpt.find_first_value_before_ts_ind(bahamas, 'roll', min_roll_after_leg1, 0., 0.01)

#radar_cal_wiggle
max_roll_before_radar_cal_wiggle = fpt.max_timestamp_ind(bahamas, 'roll',
                                   slice('2020-02-13T12:00:00', '2020-02-13T12:20:00'), order=1000)

enter_radar_cal_wiggle = fpt.find_first_value_after_ts_ind(bahamas, 'roll', max_roll_before_radar_cal_wiggle, 0., 0.3)
exit_radar_cal_wiggle = 15458

#leg2 (return NTAS)
min_roll_before_leg2 = fpt.min_timestamp_ind(bahamas, 'roll', 
                       slice('2020-02-13T12:10:00', '2020-02-13T12:30:00'), order=1000)

enter_leg2 = fpt.find_first_value_after_ts_ind(bahamas, 'roll', min_roll_before_leg2, 0., 0.05)

min_roll_after_leg2 = fpt.min_timestamp_ind(bahamas, 'roll', 
                      slice('2020-02-13T12:50:00', '2020-02-13T12:59:00'), order=1000)

exit_leg2 = fpt.find_first_value_before_ts_ind(bahamas, 'roll', min_roll_after_leg2, 0., 0.05) 

#2nd circle set
###############
enter_c4_ds = fpt.ds_timestamp_ind(bahamas, dropsondes, 37)
enter_c4 = fpt.timestamp_ind_1min_prior(bahamas, enter_c4_ds)
#exit_c4 = fpt.exit_circle_timestamp_ind(bahamas, enter_c4)
#enter_c4 = fpt.ds_timestamp_ind(bahamas, dropsondes, 37)
exit_c4  = fpt.ds_timestamp_ind(bahamas, dropsondes, 48)
print ('c4')
print ('----')
print (bahamas.heading[enter_c4_ds])
print ('----')
print (exit_c4)
print (bahamas.heading[exit_c4])
print ('----')
exit_c4 = 22100
print (exit_c4)
print (bahamas.heading[exit_c4])
print ('----')

enter_c5_ds = fpt.ds_timestamp_ind(bahamas, dropsondes, 49)
enter_c5 = fpt.timestamp_ind_1min_prior(bahamas, enter_c5_ds)
#exit_c5 = fpt.exit_circle_timestamp_ind(bahamas, enter_c5)
#enter_c5 = fpt.ds_timestamp_ind(bahamas, dropsondes, 49)
exit_c5  = fpt.ds_timestamp_ind(bahamas, dropsondes, 60)
print ('c5')
print ('----')
print (bahamas.heading[enter_c5_ds])
print ('----')
print (exit_c5)
print (bahamas.heading[exit_c5])
print ('----')
exit_c5 = 26990
print (exit_c5)
print (bahamas.heading[exit_c5])
print ('----')

enter_c6_ds = fpt.ds_timestamp_ind(bahamas, dropsondes, 61)
enter_c6 = fpt.timestamp_ind_1min_prior(bahamas, enter_c6_ds) - 100
#exit_c6 = fpt.exit_circle_timestamp_ind(bahamas, enter_c6)
#enter_c6 = fpt.ds_timestamp_ind(bahamas, dropsondes, 61)
exit_c6  = fpt.ds_timestamp_ind(bahamas, dropsondes, 72)
print ('c6')
print ('----')
print (bahamas.heading[enter_c6_ds])
print ('----')
print (exit_c6)
print (bahamas.heading[exit_c6])
print ('----')
exit_c6 = 31455
print (exit_c6)
print (bahamas.heading[exit_c6])
print ('----')
headingdiff_c6 = bahamas['heading'][exit_c6]-bahamas['heading'][enter_c6]+360


#lidar_leg
bahamas["altitude_dt"] = (['time'], np.append(np.diff(bahamas["altitude"]),0))

min_dzdt_before_lidar_leg = fpt.min_timestamp_ind(bahamas, 'altitude_dt', 
                            slice('2020-02-13T16:45:00', '2020-02-13T17:00:00'), order=1000)

enter_lidar_leg = fpt.find_first_value_after_ts_ind(bahamas, 'altitude_dt', min_dzdt_before_lidar_leg, 0., 0.05)

min_dzdt_after_lidar_leg = fpt.min_timestamp_ind(bahamas, 'altitude_dt', 
                           slice('2020-02-13T17:00:00', '2020-02-13T17:10:00'), order=1000)

exit_lidar_leg = fpt.find_first_value_before_ts_ind(bahamas, 'altitude_dt', min_dzdt_after_lidar_leg, 0., 0.05)


ts_ind_list = [takeoff, 
               enter_c1, exit_c1, 
               enter_c2, exit_c2, 
               enter_c3, exit_c3,
               enter_leg1, exit_leg1,
               enter_radar_cal_wiggle,
               exit_radar_cal_wiggle,
               enter_leg2, exit_leg2,
               enter_c4, exit_c4, 
               enter_c5, exit_c5,
               enter_c6, exit_c6,
               enter_lidar_leg, exit_lidar_leg, 
               touchdown]
label_list = ["takeoff", 
              "enter_c1", "exit_c1", 
              "enter_c2", "exit_c2", 
              "enter_c3", "exit_c3", 
              "enter_leg1", "exit_leg1",
              "enter_radar_cal_wiggle",
              "exit_radar_cal_wiggle",
              "enter_leg2", "exit_leg2",
              "enter_c4", "exit_c4", 
              "enter_c5", "exit_c5",
              "enter_c6", "exit_c6",
              "enter_lidar_leg", "exit_lidar_leg", 
              "touchdown"]
```
